Домашнє завданя 2: Списки та словники

1. Списки - базові операції

```python
#Список температур
temperatures = [22, 25, 19, 23, 26, 20, 21, 22]
```

```python
print("Перший елемент: ", temperatures[0])
print("Останній елемент: ", temperatures[-1])

#Додавання і видалення температури
temperatures.append(24)
temperatures.remove(19)

print("Довжина списку: ", len(temperatures))
print("Сума всіх температур: ", sum(temperatures))
print("Максимальна температура: ", max(temperatures))
```

2. Словники - базові операції

```python
#Словник
person = {
    "name": "Nadiia",
    "age": 32,
    "city": "Lviv"
}
```

```python
#Робота зі словником
print("Значення ключа 'name': ",person["name"])

person["age"] = 21
person["hobby"] = "programming"

print("Всі ключі: ",person.keys())
print("Всі значення: ",person.values())
```
