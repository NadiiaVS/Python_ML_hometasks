Домашнє завданя 3: Умовні конструкції (if-elif-else)

1. Основи роботи з умовними конструкціями (if-elif-else)

```python
#час варіння яєць у хвилинах
cooking_time = 5

#результат
if cooking_time < 4:
    print("Яйце рідке")
elif 4 <= cooking_time < 7:
    print("Яйце всмятку")
elif 7 <= cooking_time < 10:
    print("Яйце вкруту")
else:
    print("Яйце переварене")
```

2. Перевірка оцінки

```python
#Оцінка
grade = 72

#Виведення результату
if grade >= 90:
    print("Відмінно")
elif 75 <= grade < 90:
    print("Добре")
elif 60 <= grade < 75:
    print("Задовільно")
else:
    print("Незадовільно")
```

3. Перевірка на наявність

```python
#Список продуктів
fridge = ["молоко","яйця","масло","сир"]

#Перевірка наявності яєць
if "яйця" in fridge:
    print("Яйця є в холодильнику")
else:
    print("Яєць немає")

#Перевірка наявності хліба
if "хліб" in fridge:
    print("Хліб є в холодильнику")
else:
    print("Хліб відсутній")
```
