1. Цикл for - базовий

```python
for i in range(1,8):
    print(i)
```

2. Цикл for по списку

```python
names = ["Олег", "Анна", "Іван", "Марія"]

for name in names:
    print(f"Привіт, {name}!")
```

3. Цикл з enumerate

```python
grades = [85, 92, 78, 95, 88]

for i, grade in enumerate(grades, start=1):
    print(f"Студент {i}: {grade} балів")
```

4. Сума парних чисел

```python
total = 0

for i in range(1,21):
    if i % 2 == 0:
        total+= i

print("Сума парних чисел від 1 до 20:", total)
```
