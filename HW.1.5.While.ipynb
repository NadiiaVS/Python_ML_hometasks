1. Цикл while

```python
counter = -3

while counter <= 5:
    counter +=1
    print(counter)
```

2. List comprehension - прості операції

```python
numbers = [11, 4, 9, 16, 7, 5]

# Квадрати всіхчисел
squares = [n**2 for n in numbers]

#Тільки парні числа
even_numbers = [n for n in numbers if n % 2 == 0]

#Всі числа помножені на 10
times_ten = [n * 10 for n in numbers]

# Результати
print(squares)
print(even_numbers)
print(times_ten)
```

3. List comprehension - комбінована логіка

```python
numbers = list(range(1,11))

result = [n * 2 if n % 2 == 0 else n - 1 for n in numbers]
print(result)
```
