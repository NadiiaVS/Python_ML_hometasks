1. Функція з параметром

```python
def show_menu(dishes):
    print("=== МЕНЮ ===")
    for i, dish in enumerate(dishes, start =1):
        print(f"{i}. {dish}")
    print("============")

#виклик функції
menu = ["Суп грибний","Котлета по-київськи","Лимонад"]
show_menu(menu)
```

2. Функція з параметром

```python
def print_square(n):
    print(f"Квадрат числа {n} = {n **2}")

#виклик функції
print_square(5)
print_square(3)
print_square(8)
```

4. Функція пошуку максимумального значення

```python
def find_max(numbers):
    """
    Функція приймає список чисел і повертає найбільше число.
    """
    return max(numbers)

#виклик функції
test_nums = [15, 42, 8, 23, 56, 19]

print("Список чисел:", test_nums)
print ("Максимальне число у списку: ",find_max(test_nums))
```

5. Функція з умовами

```python
def check_grade(grade):
    """
    Перевіряє оцінку і повертає результат.
    
    Параметри:
    grade (int or float): оцінка студента
    
    Повертає:
    str: "Здав" якщо оцінка >= 60, інакше "Не здав"
    """
    if grade >= 60:
        return "Здав"
    else:
        return "Не здав"
        
#виклик функції
print(check_grade(65))
print(check_grade(50))
print(check_grade(89))
```
